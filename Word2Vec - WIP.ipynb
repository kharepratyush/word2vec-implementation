```python
import pandas as pd
import numpy as np

data = pd.read_csv("spam.csv", encoding = "latin-1")
data = data.dropna(how = "any", axis = 1)
data.columns = ['label','body_text']

data.label = data.label.apply(lambda x: 1 if x == "spam" else 0)
data.head()
```

```python
data.shape

x_train = data.loc[:4499, 'body_text'].values
y_train = data.loc[:4499, 'label'].values

x_test = data.loc[4500:, 'body_text'].values
y_test = data.loc[4500:, 'label'].values
```

```python
x_train, x_test
```

```python
import tensorflow as tf
import numpy as np

from tensorflow.python.keras.preprocessing.text import Tokenizer
from tensorflow.python.keras.preprocessing.sequence import pad_sequences

tokenizer_obj = Tokenizer()
total_text = np.concatenate((x_train, x_test), axis=None)
total_text
```

```python
tokenizer_obj.fit_on_texts(total_text)

max_length = max([len(s.split()) for s in total_text])

vocab_size = len(tokenizer_obj.word_index) + 1

x_train_token = tokenizer_obj.texts_to_sequences(x_train)
x_test_token = tokenizer_obj.texts_to_sequences(x_test)

x_train_pad = pad_sequences(x_train_token, max_length, padding = 'post')
x_test_pad = pad_sequences(x_test_token, max_length, padding = 'post')
```
